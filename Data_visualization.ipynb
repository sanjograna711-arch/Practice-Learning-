```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('/content/spotify.csv',index_col="Date", parse_dates = True)
df.head()
df.tail()
```

```python

plt.figure(figsize=(14,6))
sn.lineplot(data = df)
```

```python

plt.figure(figsize=(14,6))
sn.lineplot(data = df[['Shape of You', 'Despacito', 'Something Just Like This']])
```

```python
flight_path = ("/content/flight_delays.csv")
flight_data = pd.read_csv(flight_path,index_col= 'Month')
```

```python
flight_data.columns
```

```python
plt.figure(figsize = (12,6))
sn.barplot(x = flight_data.index, y = flight_data["NK"],color = "#d95f5f")
```

```python

plt.figure(figsize = (10,6))

sn.heatmap(data=flight_data , annot = True)

plt.title("Flight delay heatmap")
plt.xlabel("Flight name")

plt.show()
```

```python
insurance_data = pd.read_csv("/content/insurance.csv")
insurance_data.columns
```

```python
plt.figure(figsize = (10,6))

sn.regplot(x = insurance_data.bmi , y = insurance_data.charges,color = "#f065f7")
sn.scatterplot(x = insurance_data.bmi , y = insurance_data.charges ,hue = insurance_data.age)

plt.title("Relationship between BMI, Age and Charges")

plt.show()
```

```python
plt.figure(figsize = (16,5))

sn.lmplot(x = "bmi", y = "charges" , hue = "smoker", data= insurance_data)

plt.show()
```

```python

plt.figure(figsize = (10,10))

sn.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['bmi'],hue = insurance_data.age)

plt.show()
```

```python
iris_data = pd.read_csv("/content/iris.csv",index_col = "Id")
iris_data.columns
```

```python

plt.figure(figsize = (16,8))
sn.histplot(data = iris_data ["Petal Length (cm)"])

plt.title("Petal Length (cm)")

plt.show()
```
